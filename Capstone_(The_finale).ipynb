```python
#Removes the "\" from the player name
def Player_name_fix(ht):
    #Fixes Player name to remove "\"
    ht_ = ht.split("\\")
    KeptPortion = ht_[0]
    return KeptPortion

#Changes height from feet into inches
def feet_to_inches(ht):
    #Split the height between ft and inches
    ht_ = ht.split("-")
    #Convert to float
    ft_ = float(ht_[0])
    in_ = float(ht_[1])
    ft_to_in = (12*ft_) + in_
    return ft_to_in
```

```python
df["Height"] = df["Height"].apply(feet_to_inches)
```

Scrapes data from site and places it into a dataframe

```python
#https://www.youtube.com/watch?v=ooj84UP3r6M - By the Dataprofessor
#Takes the URL of the site where data was taken
import pandas as pd
year = 2000

site = "https://www.pro-football-reference.com/years/{}/draft.htm"

url = site.format(year)
url
```

```python
#Turns site info into list
df_scraping_test = pd.read_html(url, header = [1])
#Turns list into dataframe, remove other tables not needed
df_scraping_test = df_scraping_test[0]

df_scraping_test
```

```python
#While loop to scrape data and place into a single CSV file
i = 0
while i < 18:
    url = site.format(year+i)
    #Turns site info into list
    df_scraping_test1 = pd.read_html(url, header = [1])
    #Turns list into dataframe, remove other tables not needed
    df_scraping_test1 = df_scraping_test1[0]
    i+=1
    df_scraping_test = df_scraping_test.append(df_scraping_test1, ignore_index=True)

df_scraping_test.to_csv('draft'+str(year)+'.csv', index=False)
```

Merging dataframes

```python
#Creation of a segment for people who were drafted
df_scraping_segment = df_scraping_test[['Player','Pos',]]
df_scraping_segment
```

```python
import numpy as np
df_scraping_segment["Drafted"] = "True"
df_scraping_segment
```

```python
df["Player"] = df["Player"].apply(Player_name_fix)
df
```

```python
#Combining dataframes of scraped dataset and adjusted combine set
#Used outer join to distigush between those drafted or not
mergedDataFrame = pd.merge(df_scraping_segment, df, on='Player', how='right')
mergedDataFrame
```

```python
#Since data dat empty values in the drafted category fill with false
mergedDataFrame[['Drafted']] = mergedDataFrame[['Drafted']].fillna(value="False")
mergedDataFrame
```

```python
mergedDF = mergedDataFrame.drop_duplicates(subset=['Player'])
mergedDF
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Replace zeroes
zero_not_accepted = ['Height', 'Wt', '40YD', 'Vertical', 'BenchReps','Broad Jump', '3Cone', 'Shuttle']
```

```python
for column in zero_not_accepted:
    mergedDF[column] = mergedDF[column].replace(0, np.NaN)
    mean = int(mergedDF[column].mean(skipna=True))
    mergedDF[column] = mergedDF[column].replace(np.NaN, mean)

mergedDF
```

```python
mergedDF = mergedDF.drop(['School', 'College', 'Drafted (tm/rnd/yr)', 'AV', 'Pos_x'], axis=1)
mergedDF
```

```python
#Replacing draft values with numbers
mergedDF = mergedDF.replace(to_replace =["True", "False"],
                            value =["1", "2"])
```

```python
# https://www.youtube.com/watch?v=4HKqjENq9OU&t=590s
#Testing the values of the player attributes (Takes all rows and takes columns 10-17)
X = mergedDF.iloc[:, 5:13]
#Determing whether or not the player was drafted
y = mergedDF.iloc[:, 1]
#Testing the set, setting aside 20% of the values for later
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
# Define the model: Init K-NN
classifier = KNeighborsClassifier(n_neighbors=20, p=1,metric='euclidean')
```

```python
# Fit Model
classifier.fit(X_train, y_train)
```

```python
#Represents the number of values to be tested and trained
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python
# Predict the test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# Evaluate Model using confusion matrix
cm = confusion_matrix(y_test, y_pred)
print (cm)
```

```python
#This shows the accuracy of the model in predicting whether or not a player will be drafted
print(str(accuracy_score(y_test, y_pred)*100)+'%')
```

```python
#Indication of false positives in the model
print(f1_score(y_test, y_pred, average=None))
```
